```python
from googletrans import Translator
import pandas as pd

translator = Translator()
# idx = [13, 18, 19]
for i in range(0, 20):
    input_string = df3.iloc[i, 1]
    output_string = translator.translate(input_string, dest='ko', src='en')
    df3.iloc[i, 1] = output_string.text
df3.to_csv('translated_pop_music_lyrics.csv', index=False)
```
